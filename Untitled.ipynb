```matlab
clear 
close all


[s_real,s_imaginaria] = meshgrid([-5:0.3:5],[-20:0.5:20]);
s = s_real+j*s_imaginaria; %Definicin de la variable compleja s=real+imaginaria
Xs = @(s) (1)./(2.5*s.^2+3*s+350);  % Transformada de Laplace de la salida de un sistema
MagXs = abs(Xs(s));

% Se define un rango de frecuencias de inters.
omega = [-25:0.1:25];
tr = j*omega;
% Se produce una retcula para generar la superficie
%shading interp;
%colormap copper;
m1 = mesh(s_real,s_imaginaria,MagXs);
set(gcf,'Visible','on')
hold on;
% Se superpone la curva de magnitud de X(s)
%  usando el comando plot3().
m2 = plot3(real(tr),imag(tr),abs(Xs(tr)),'b','LineWidth',1.5);
hold off;
axis([-5,5,-25,25]);
% Se especifican los rtulos para los ejes x,y y z.
xlabel('\sigma');
ylabel('j\omega');
zlabel('|X(s)|');
% Se especifica un ngulo de mejor visualizacin. 
view(gca,[23.5,38]);
set(gca,'fontsize',20);
%set(gca,'fontname','Times New Roman','FontWeight','Bold'); 
```
